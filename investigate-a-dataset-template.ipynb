## Data cleaning

>1)Remove unused columns

>2)Remove duplicate rows if found

>3)Rchange the zero values to Nan and remving

>4)Change the formatting of some columns



## 1)Remove unused columns

```python
#creating a list to delete the coulmns we not need
delete_col=[ 'id', 'imdb_id', 'popularity', 'budget_adj', 'revenue_adj', 'homepage', 'keywords', 'overview', 'production_companies', 'vote_count', 'vote_average']

#deleting the columns
df= df.drop(delete_col,1)

#printing the data
df.head(10)
```

## 2) Remove duplicate rows if found



```python
rows, columns = df.shape
print('The Total is {} and Noumber of Columns it is {}'.format(rows-1, columns))
```

```python
#Now deleting duplicates row
df.drop_duplicates(keep ='first', inplace=True)
rows, columns = df.shape
#print the number of columns 
print('The Total is {} and Noumber of Columns It Is {}'.format(rows-1, columns))
```

## 3)change the zero values to Nan and remving

```python
# creating a data list of revenue and budget column to remove zero
data_list=['budget', 'revenue']

#change value from 0 to NAN
df[data_list] = df[data_list].replace(0, np.NAN)

#replace the colmn runtime value zero to NAN 
df['runtime'] =df['runtime'].replace(0, np.NAN)

#now we will deleting all the row that have NaN value in data_list 
df.dropna(subset = data_list, inplace = True)

rows, column = df.shape
print('Now we have only {} of movies.'.format(rows-1))
```

## 5)Change the formatting of some columns

```python
#print the data set type
df.dtypes
```

```python
df.release_date = pd.to_datetime(df['release_date'])
change_format=['budget', 'revenue']
#Now changing the format of data set ('budget', 'revenue')
df[change_format]=df[change_format].applymap(np.int64)
#printing the data set after change
df.dtypes
df.head(10)
```

### Research Question 2  (what the high and low profits in movies!)

```python
#Calculate profit
df.insert(2,'profit',df['revenue']-df['budget'])

#printing the data set after change
df.head(5)
```

Add a column to calculate the profit

```python
#defining the function
def calcultion(col):
    #for highest earned profit
    high_profit= df[col].idxmax()
    high=pd.DataFrame(df.loc[high_profit])
    
    #for lowest earned profit
    low_profit= df[col].idxmin()
    low=pd.DataFrame(df.loc[low_profit])
    
    #collectin data in one place
    info=pd.concat([high, low], axis=1)
    
    return info

#calling the function
calcultion('profit')
```

The film's high profit is Avatar and the least profitable film is The Warrior's Way

### Research Question 3  (genres Vs. profits of movies!)

```python
profits_Of_Years= df.groupby('genres')['profit'].sum()

plt.figure(figsize=(20,5), dpi = 75)

plt.xlabel('genres of Movies in the data set', fontsize = 12)

plt.ylabel('Profits ', fontsize = 12)
#title of histogram
plt.title('Profits of all movies Vs genres')

plt.plot(profits_Of_Years)

plt.show()

```

The most profitable film it is genre from action
